```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import skew
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
plt.rcParams['figure.figsize']=(12,8)
```

```python
data=pd.read_csv("car.csv")
```

```python
data.head()
```

```python
x=np.array(data[['year']])
y=np.array(data['sellingprice'])
```

```python
plt.scatter(x,y)
plt.xlabel('year')
plt.ylabel('sellingprice')
plt.title('year vs sellingprice')
plt.show()
```

```python
x_train=x[0:120]
y_train=y[0:120]
x_test=x[120:]
y_test=y[120:]
```

```python
x_mean= np.mean(x_train)
y_mean= np.mean(y_train)
print(x_mean,y_mean)
```

```python
num = 0
deno = 0

for i in range(0,len(x_train)):
  num += (x[i]-x_mean)*(y[i]-y_mean)
  deno += (x[i]-x_mean)**2

m = num/deno

print(m)
```

```python
c=y_mean-(m*x_mean)
print(c)
```

```python
y_pred=m*x_test+c
print(y_pred)
```

```python
for i in range(0,len(x_test)):
  print("Actual Y: " + str(y_test[i]) + "  Predicted Y " + str(y_pred[i]))
```
